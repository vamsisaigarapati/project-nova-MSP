```python
from src.configs.hearst_configs import raw_column_types
import pandas as pd
from pathlib import Path

excel_path = Path(HEARST_RAW_DIR) / "Hearst Files.xlsx"
dtype_dict, parse_dates, converters = {}, [], {}

for d in raw_column_types:
    (col, dtype), = d.items()
    if dtype == "datetime64[ns]":
        parse_dates.append(col)
    elif dtype is int:
        # per-cell: coerce to number; returns scalar (float or NaN)
        converters[col] = lambda v: pd.to_numeric(v, errors="coerce")
    elif dtype is float:
        converters[col] = lambda v: pd.to_numeric(v, errors="coerce")
    else:
        dtype_dict[col] = dtype  # str/object handled here

# read — note sheet name "Raw"
raw_df = pd.read_excel(
    excel_path,
    sheet_name="Raw",
    dtype=dtype_dict,
    parse_dates=parse_dates,
    converters=converters,
)

# final, column-level cast to nullable ints for the int columns
for d in raw_column_types:
    (col, dtype), = d.items()
    if dtype is int and col in raw_df.columns:
        raw_df[col] = pd.to_numeric(raw_df[col], errors="coerce").astype("Int64")

print("✅ Loaded with safe conversions.")
print(raw_df.dtypes.head(20))
print(raw_df.head())
```

```python
import pandas as pd
import os
from pathlib import Path

def calculate_revenue(raw_df):
    """
    Reads the 'Hearst Pub Market List' sheet from 'Heast Files.xlsx',
    joins with raw_df on 'Pub', creates 'Job Number +',
    and returns a DataFrame grouped by 'Job Number +' with summed Revenue.
    """
    # Build Excel file path
    excel_path = Path(os.path.join(HEARST_RAW_DIR, "Heast Files.xlsx"))
    if not excel_path.exists():
        excel_path = Path(os.path.join(HEARST_RAW_DIR, "Hearst Files.xlsx"))

    # Read the "Hearst Pub Market List" sheet
    market_list = pd.read_excel(excel_path, sheet_name="Hearst Pub Market List")

    # Prepare join keys (case-insensitive and trimmed)
    raw_df = raw_df.copy()
    market_list = market_list.copy()
    raw_df["Pub_key"] = raw_df["Pub"].astype(str).str.strip().str.lower()
    market_list["Pub_key"] = market_list["Pub"].astype(str).str.strip().str.lower()

    # Merge Market info
    merged_df = raw_df.merge(
        market_list[["Pub_key", "Market"]],
        on="Pub_key",
        how="inner"
    )

    # Create "Job Number +" column = Market + Job Number
    merged_df["Job Number"] = merged_df["Job Number"].astype(str).str.strip()
    merged_df["Market"] = merged_df["Market"].astype(str).str.strip()

    merged_df["Job Number +"] = merged_df.apply(
        lambda r: f"{r['Market']}{r['Job Number']}"
        if r["Market"] not in ["", "nan", "None"]
        else r["Job Number"],
        axis=1
    )

    # Convert Revenue to numeric and sum
    merged_df["Revenue"] = pd.to_numeric(merged_df["Revenue"], errors="coerce").fillna(0)

    result_df = (
        merged_df
        .groupby("Job Number +", as_index=False)["Revenue"]
        .sum()
    )

    return result_df
```

```python
import pandas as pd
import os
from pathlib import Path

def calculate_revenue(raw_df):
    """
    Reads the 'Hearst Pub Market List' sheet from 'Heast Files.xlsx',
    joins with raw_df on 'Pub', creates 'Job Number +',
    and returns a DataFrame grouped by 'Job Number +' with summed Revenue.
    Output columns: ['Old Job Number', 'Job Number', 'Job Number +', 'Revenue'].
    """
    # Build Excel file path
    excel_path = Path(os.path.join(HEARST_RAW_DIR, "Heast Files.xlsx"))
    if not excel_path.exists():
        excel_path = Path(os.path.join(HEARST_RAW_DIR, "Hearst Files.xlsx"))

    # Read the "Hearst Pub Market List" sheet
    market_list = pd.read_excel(excel_path, sheet_name="Hearst Pub Market List")

    # Prepare join keys (case-insensitive and trimmed)
    merged_df = raw_df.copy()
    market_list = market_list.copy()
    merged_df["Pub_key"] = merged_df["Pub"].astype(str).str.strip().str.lower()
    market_list["Pub_key"] = market_list["Pub"].astype(str).str.strip().str.lower()

    # Merge Market info (inner to match your previous behavior)
    merged_df = merged_df.merge(
        market_list[["Pub_key", "Market"]],
        on="Pub_key",
        how="inner"
    )

    # Normalize columns for concatenation
    merged_df["Job Number"] = merged_df["Job Number"].astype(str).str.strip()
    merged_df["Market"] = merged_df["Market"].astype(str).str.strip()

    # Create "Job Number +" = Market + Job Number (no separator, as before)
    merged_df["Job Number +"] = merged_df.apply(
        lambda r: f"{r['Market']}{r['Job Number']}"
        if r["Market"] not in ["", "nan", "None"]
        else r["Job Number"],
        axis=1
    )

    # Convert Revenue to numeric
    merged_df["Sum of 'Revenue'"] = pd.to_numeric(merged_df["Revenue"], errors="coerce").fillna(0)


    # Group by 'Job Number +' and sum revenue; keep representative originals
    
    result_df = (
        merged_df
        .groupby("Job Number +", as_index=False)
        .agg({
            "Sum of 'Revenue'": "sum",
            "Job Number": "first"
        })
    )

    # Reorder columns
    result_df = result_df[["Job Number", "Job Number +", "Sum of 'Revenue'"]]

    return result_df
```

```python
import pandas as pd
import os
from pathlib import Path

def calculate_revenue(raw_df):
    """
    Reads the 'Hearst Pub Market List' sheet from 'Heast Files.xlsx',
    joins with raw_df on 'Pub', creates 'Job Number +',
    and returns a DataFrame grouped by 'Job Number +' with:
      - the first encountered record for all columns,
      - summed Revenue,
      - and count of records in each group ('Count of Matches').
    """
    # Build Excel file path
    excel_path = Path(os.path.join(HEARST_RAW_DIR, "Heast Files.xlsx"))
    if not excel_path.exists():
        excel_path = Path(os.path.join(HEARST_RAW_DIR, "Hearst Files.xlsx"))

    # Read the "Hearst Pub Market List" sheet
    market_list = pd.read_excel(excel_path, sheet_name="Hearst Pub Market List")

    # Prepare join keys (case-insensitive and trimmed)
    merged_df = raw_df.copy()
    market_list = market_list.copy()
    merged_df["Pub_key"] = merged_df["Pub"].astype(str).str.strip().str.lower()
    market_list["Pub_key"] = market_list["Pub"].astype(str).str.strip().str.lower()

    # Merge Market info (inner join for exact matches)
    merged_df = merged_df.merge(
        market_list[["Pub_key", "Market"]],
        on="Pub_key",
        how="inner"
    )

    # Normalize columns for concatenation
    merged_df["Job Number"] = merged_df["Job Number"].astype(str).str.strip()
    merged_df["Market"] = merged_df["Market"].astype(str).str.strip()

    # Create "Job Number +" = Market + Job Number
    merged_df["Job Number +"] = merged_df.apply(
        lambda r: f"{r['Market']}{r['Job Number']}"
        if r["Market"] not in ["", "nan", "None"]
        else r["Job Number"],
        axis=1
    )

    # Convert Revenue safely to numeric
    merged_df["Sum of 'Revenue'"] = pd.to_numeric(merged_df["Revenue"], errors="coerce").fillna(0)

    # ---- Aggregate ----
    # Keep first record for every column except the revenue column
    agg_dict = {col: "first" for col in merged_df.columns if col not in ["Sum of 'Revenue'"]}
    agg_dict["Sum of 'Revenue'"] = "sum"

    result_df = merged_df.groupby("Job Number +", as_index=False).agg(agg_dict)

    # Add the count of records per group
    counts = merged_df.groupby("Job Number +").size().reset_index(name="Count of Matches")
    result_df = result_df.merge(counts, on="Job Number +", how="left")

    # Reorder columns — keep Job Number +, Sum of Revenue, Count of Matches at the end
    cols = [c for c in result_df.columns if c not in ["Job Number +", "Sum of 'Revenue'", "Count of Matches"]]
    result_df = result_df[cols + ["Job Number +", "Sum of 'Revenue'", "Count of Matches"]]

    return result_df
```
